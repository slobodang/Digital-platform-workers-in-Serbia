```python
#Getting the freelancer profile links with Selenium

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

options = Options()#Installing webdriver options
chrome_driver = webdriver.Chrome(chrome_options=options)#starting a webdriver
chrome_driver.get('https://www.freelancer.com/freelancers/serbia')#initial page
chrome_driver.implicitly_wait(10)#random implicit waits
chrome_driver.find_element_by_id("selected-online").click()#deselecting the online button
chrome_driver.implicitly_wait(10)

profile_links = []#getting the profile links in a list

condition = True

while condition:
    chrome_driver.implicitly_wait(10)
    try:
        profiles_list = chrome_driver.find_elements_by_class_name("freelancer-details")#profile locations
        for profile in profiles_list:
            chrome_driver.implicitly_wait(10)
            x1 = profile.find_element_by_tag_name('h3')#finding link in html
            chrome_driver.implicitly_wait(10)
            x2 = x1.find_element_by_tag_name('a')#finding link in html
            chrome_driver.implicitly_wait(10)
            profile_links.append(x2.get_attribute('href'))#appending href links to the list
    except (StaleElementReferenceException, NoSuchElementException) as Exception:#most common exceptions
        print(Exception)
    pass
    try:
        chrome_driver.implicitly_wait(10)
        chrome_driver.find_element_by_css_selector('#display_div > div:nth-child(1) > div.ns_pagination.is-desktop-only > ul > li:nth-child(7)').click()
    except: #finding and selecting the next page button
        condition = False
#No idea how to solve the problems I have, except for extension of waiting times till the elements are found...
```

```python
#Parsing individual profiles with BeautifulSoup
from bs4 import BeautifulSoup
soups = []#list of soup files
folderpath = r'C:\Users\Slobodan\Desktop\CEU PhD\Digital data collection methods\final project\freelancer_profiles_selenium'
for filename in os.listdir(folderpath):
    fname = os.path.join(folderpath,filename)
    with open(fname, 'r', encoding = 'utf-8') as f:
        soup = BeautifulSoup(f.read(),'html.parser')#parsing the profiles
        soups.append(soup)
          
```

```python
print(len(soups))
```

```python
#The main loop. Scraping the variables of interest and placing them inside a json file

#creating an empty json file for profiles
import json
json_profiles = {}
json_profiles['profiles'] = []

#making a large loop to iterate through all the variables I am interested in
for soup in soups:
    try:#using this because one of the hourly rates has a different code on few profiles for some reason
        #finding profile unique names
        name_parse = soup.find_all('h3')[-1]#locating the id name
        name_string = name_parse.get_text()#getting the text from the id paragraph
        name = name_string[1:]#eliminating the @ sign before the name
        print(name)
        
        #finding user hourly rate
        find_hourly_rate = soup.find_all('fl-col')[1]#finding the second column on basis of html inspection
        hourly_rate = find_hourly_rate.find_all(class_ = 'NativeElement ng-star-inserted')[1]# finding the second class
        rate_text = hourly_rate.get_text()#it is in the form of text
        import re
        rate = re.findall(r'\d+', rate_text)#selecting numbers
        rate = int(rate[0])#taking the number out of the list and converting it to int
        print(rate)
        
        #finding skills for each profile
        find_skills = soup.find_all(class_ = 'UserProfileSkill')#locating the unique skill class
        skills = []#list of skills with their frequency of use for jobs for each profile
        for skill in find_skills:
            skill_item = skill.get_text()#getting text out of this script
            skill_item = skill_item.replace('\n', '')#eliminating frequent \n signs 
            skills.append(skill_item)
        print(skills)
        
        #finding employer countries  #PROBLEMS WITH THIS!
        find_employers = soup.find_all('fl-list')[0]#it's the first fl-list in the html file
        employers = find_employers.find_all(class_ = 'FlagImage')#'only member of this class'
        employer_countries = []#list of employer countries for each profile
        for country in employers:
            country = str(country)[-5:-3]#selecting only the country initials from the script
            employer_countries.append(country)
        print(employer_countries)
        
        #appending each of the variables from the loop to a key in json object
        json_profiles['profiles'].append({
        'id' : name,
        'hourly_rate' : rate,
        'top_skils' : skills,
        'employers_countries' : employer_countries
    })
    except:
        pass
    
```

```python
#Converting json file to pandas. I assume I can analyze them even better in json, but I am just more comfortable with pandas
import pandas as pd
data = json_profiles['profiles']

dictionary = [i for i in data[0] if type(data[0][i]) != dict]

dic = {i:[] for i in dictionary}
for i in dictionary:
    for j in range(len(data)):
        dic[i].append(data[j][i])
        

pd.set_option('display.max_rows', None)#displaying all of the rows

df_profiles = pd.DataFrame(dic)
df_profiles#two of the profiles were omitted. I might work on that
```

```python
#Calculating employer country frequencies

#converting two letter country codes to full names
!pip install country-converter
import country_converter as coco

#counting most common employer countries accross the sample
from collections import Counter
country_counter = Counter(x for sub_list in df_profiles_jobs['employers_countries'] for x in sub_list)
most_common_employers = country_counter.most_common()
most_common_employers

#making country codes mutable
most_common_employer_countries = []
for tup in most_common_employers:
    most_common_employer_countries.append(list(tup))
    
#converting country codes to full names
for x in most_common_employer_countries:
    x[0] = coco.convert(names = x[0], to = 'name_short')
most_common_employer_countries

#I understand that there are much more elegant ways to do this, but I figured out that I need full names for graphs quite late#
#And it is difficult for me now to change this in pandas, or earlier in the scrapping process.
```

```python
#Visualizing the most common employer countries on the total sample

import numpy as np
import matplotlib.pyplot as plt

#selecting only top 10 employer countries
top_10_employers = most_common_employer_countries[:10]
print(top_10_employers)

#making my pairs suitable for a plot
n_groups = len(top_10_employers)

frequencies = [x[1] for x in top_10_employers]
countries = [x[0] for x in top_10_employers]

#Plotting

fig_employers, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.4

opacity = 0.6

rects1 = plt.bar(index, frequencies, bar_width,
                 alpha=opacity,
                 color='r', label = 'Frequency')


plt.xlabel('Employer country')
plt.ylabel('Frequency')
plt.title('Most frequent employer countries')
plt.xticks(index + bar_width, countries, rotation = 60)
plt.legend()

plt.tight_layout()
plt.show()

fig_employers.savefig('employers_plot.png')


```

```python
#creating a function which would automatize visualization of emloyer country frequencies accross professions dataframes

def visualize_country_frequencies(data_frame, title):
    country_counter_ = Counter(x for sub_list in data_frame['employers_countries'] for x in sub_list)
    most_common_employers_ = country_counter_.most_common()
    

    #making country codes mutable
    most_common_employer_countries_ = []
    for tup in most_common_employers_:
        most_common_employer_countries_.append(list(tup))
    
    #converting country codes to full names
    for x in most_common_employer_countries_:
        x[0] = coco.convert(names = x[0], to = 'name_short')
        most_common_employer_countries_

    top_10_employers_ = most_common_employer_countries_[:10]
    print(top_10_employers_)

    n_groups_ = len(top_10_employers_)
    index_ = np.arange(n_groups_)

    frequencies_ = [x[1] for x in top_10_employers_]
    countries_ = [x[0] for x in top_10_employers_]


    fig_employers_, ax_ = plt.subplots()

    bar_width_ = 0.4

    opacity_ = 0.6

    rects1_ = plt.bar(index_, frequencies_, alpha=opacity_, color='r', label = 'Frequency')

    plt.xlabel('Employer country')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(index_, countries_, rotation = 60)
    plt.legend()

    plt.tight_layout()
    plt.show()
    return



```
