```python
#Getting the soup files for individual profiles
import os
soups = []#list of soup files
folderpath = r'C:\Users\Slobodan\Desktop\CEU PhD\Digital data collection methods\final project\100profiles'
for filename in os.listdir(folderpath):
    fname = os.path.join(folderpath,filename)
    with open(fname, 'r', encoding = 'utf-8') as f:
        soup = BeautifulSoup(f.read(),'html.parser')#parsing the profiles
        soups.append(soup)
          
```

```python
#The main loop. Scraping the variables of interest and placing them inside a json file

#creating an empty json file for profiles
import json
json_profiles = {}
json_profiles['profiles'] = []

#making a large loop to iterate through all the variables I am interested in
for soup in soups:
    try:#using this because one of the hourly rates has a different code on few profiles for some reason
        #finding profile unique names
        name_parse = soup.find_all('h3')[-1]#locating the id name
        name_string = name_parse.get_text()#getting the text from the id paragraph
        name = name_string[1:]#eliminating the @ sign before the name
        print(name)
        
        #finding user hourly rate
        find_hourly_rate = soup.find_all('fl-col')[1]#finding the second column on basis of html inspection
        hourly_rate = find_hourly_rate.find_all(class_ = 'NativeElement ng-star-inserted')[1]# finding the second class
        rate_text = hourly_rate.get_text()#it is in the form of text
        import re
        rate = re.findall(r'\d+', rate_text)#selecting numbers
        rate = int(rate[0])#taking the number out of the list and converting it to int
        print(rate)
        
        #finding skills for each profile
        find_skills = soup.find_all(class_ = 'UserProfileSkill')#locating the unique skill class
        skills = []#list of skills with their frequency of use for jobs for each profile
        for skill in find_skills:
            skill_item = skill.get_text()#getting text out of this script
            skill_item = skill_item.replace('\n', '')#eliminating frequent \n signs 
            skills.append(skill_item)
        print(skills)
        
        #finding employer countries
        find_employers = soup.find_all('fl-list')[0]#it's the first fl-list in the html file
        employers = find_employers.find_all(class_ = 'FlagImage')#'only member of this class'
        employer_countries = []#list of employer countries for each profile
        for country in employers:
            country = str(country)[-5:-3]#selecting only the country initials from the script
            employer_countries.append(country)
        print(employer_countries)
        
        #appending each of the variables from the loop to a key in json object
        json_profiles['profiles'].append({
        'id' : name,
        'hourly_rate' : rate,
        'top_skils' : skills,
        'employers_countries' : employer_countries
    })
    except:
        pass
```

```python
#Converting json file to pandas
data = json_profiles['profiles']

dictionary = [i for i in data[0] if type(data[0][i]) != dict]

dic = {i:[] for i in dictionary}
for i in dictionary:
    for j in range(len(data)):
        dic[i].append(data[j][i])
        

pd.set_option('display.max_rows', None)#displaying all of the rows

df_profiles = pd.DataFrame(dic)
df_profiles#two of the profiles were omitted. I might work on that
```
